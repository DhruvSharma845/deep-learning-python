```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
import os
DATA_BASE_PATH = 'datasets/climate'
DATA_FILE_NAME = 'jena_climate_2009_2016.csv'
climate_df = pd.read_csv(os.path.join(os.curdir, DATA_BASE_PATH, DATA_FILE_NAME))
climate_df.info()
```

```python
climate_df.drop(labels=['Date Time'], axis = 1, inplace=True)
climate_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

vals = climate_df.values
std_scaler = StandardScaler()
vals_scaled = std_scaler.fit_transform(vals)
climate_df = pd.DataFrame(vals_scaled, columns=climate_df.columns, index=climate_df.index)
climate_df.head()
```

```python
target_df = climate_df[['T (degC)']]
target_df.info()
```

```python
climate_df = climate_df.drop(['T (degC)'], axis=1)
climate_df.info()
```

```python
NUM_DAYS_PAST_LOOKUP = 10
# Instead of every 10 min, take obervations fron every hour
STEP = 6 
ENTRIES_IN_A_DAY = 6 * 24

# 1440 entries in a window
ENTRIES_IN_A_WINDOW =  ENTRIES_IN_A_DAY * NUM_DAYS_PAST_LOOKUP

# Take target entry one day ahead
LOOKAHEAD_TARGET_ENTRY = 24 * 6

def sequence_generator(features, targets, min_index, max_index, batch_size = 32):
    current_index = min_index + ENTRIES_IN_A_WINDOW
    while 1:
        if current_index + batch_size >= max_index:
            current_index = min_index + ENTRIES_IN_A_WINDOW

        rows = np.arange(current_index, min(current_index + batch_size, max_index))
        #print('Current Index: {} Max CurrentIndex: {}'.format(current_index, min(current_index + batch_size, max_index)))
        current_index += len(rows)

        samples = np.zeros((len(rows), ENTRIES_IN_A_WINDOW//STEP, len(features.columns)))
        target = np.zeros((len(rows),))

        for j, row in enumerate(rows):
            indices = range(rows[j] - ENTRIES_IN_A_WINDOW, rows[j], STEP)
            samples[j] = features.iloc[indices].values
            target[j] = targets.iloc[rows[j] + LOOKAHEAD_TARGET_ENTRY].values
        yield samples, target
```

```python
BATCH_SIZE = 32

train_generator = sequence_generator(climate_df, target_df, 0, 300000, BATCH_SIZE)
val_generator = sequence_generator(climate_df, target_df, 300001, 370000, BATCH_SIZE)
test_generator = sequence_generator(climate_df, target_df, 370001, 420000, BATCH_SIZE)

train_steps = ((300000 - ENTRIES_IN_A_WINDOW - LOOKAHEAD_TARGET_ENTRY) // BATCH_SIZE) - 1
val_steps = ((370000 - 300001 - ENTRIES_IN_A_WINDOW - LOOKAHEAD_TARGET_ENTRY) // BATCH_SIZE)
test_steps = ((420000 - 370001 - ENTRIES_IN_A_WINDOW - LOOKAHEAD_TARGET_ENTRY) // BATCH_SIZE) - 1
print('Steps: Train({}), Validation({}), Test({})'.format(train_steps, val_steps, test_steps))
```

```python
network = tf.keras.models.Sequential([
    tf.keras.layers.Conv1D(32, 5, activation='relu', input_shape=(None,len(climate_df.columns))),
    tf.keras.layers.MaxPooling1D(2),
    tf.keras.layers.GRU(64, recurrent_dropout=0.5, return_sequences=True),
    tf.keras.layers.GRU(64, recurrent_dropout=0.5, return_sequences=True),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(32)),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(1))
])

network.compile(optimizer='rmsprop', loss='mae')
network.summary()
```

```python
history = network.fit_generator(train_generator, steps_per_epoch=train_steps, epochs=1, validation_data=val_generator, validation_steps=val_steps)
history.history
```
